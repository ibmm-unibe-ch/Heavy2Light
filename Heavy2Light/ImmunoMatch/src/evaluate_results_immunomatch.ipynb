```python
import pandas as pd
import re

# Read CSV
df = pd.read_csv('/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2GPT/multiple_light_seqs_from_single_heavy/full_test_set_multiple_light_seqs/pairing_result_matching_seqs_multiple_light_seqs_203276_cls_predictions_parsed_reformatted_rel_cols.csv')

# Function to extract gene parts
def extract_gene_parts(gene_str):
    if pd.isna(gene_str):
        return None, None
    # Extract part before '*' or space
    simple = re.split(r'[\*\s]', str(gene_str))[0]
    # Extract family (remove last part after last hyphen)
    family = re.sub(r'-[^-]*$', '', simple)
    return simple, family

# Apply to true_light_first_hit_gene
df['true_v_gene_simple'], df['true_v_gene_family_call'] = zip(*df['true_light_first_hit_gene'].apply(extract_gene_parts))

# Apply to gen_light_1_first_hit_gene  
df['gen_v_gene_simple'], df['gen_v_gene_family_call'] = zip(*df['gen_light_1_first_hit_gene'].apply(extract_gene_parts))


# Save result
#df.to_csv('/ibmm_data2/oas_database/paired_lea_tmp/paired_model/BERT2GPT/multiple_light_seqs_from_single_heavy/full_test_set_multiple_light_seqs/pairing_result_matching_seqs_multiple_light_seqs_203276_cls_predictions_parsed_reformatted_rel_cols_v_genes.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import colorcet as cc
import pandas as pd

palette = sns.color_palette(cc.glasbey, n_colors=22)

plt.rcParams.update({
    'axes.titlesize': 26,
    'axes.labelsize': 22,
    'xtick.labelsize': 22,
    'ytick.labelsize': 22,
    'legend.fontsize': 18,        # Legend values font size
    'legend.title_fontsize': 18   # Legend title font size
})

# 1. Define B cell types
merged['B_cell_type'] = 'Other'
naive_condition = (merged['predicted_gen_light_seq_label'] == 0) & (merged['predicted_input_heavy_seq_label'] == 0)
memory_condition = (merged['predicted_gen_light_seq_label'] == 1) & (merged['predicted_input_heavy_seq_label'] == 1)

merged.loc[naive_condition, 'B_cell_type'] = 'Naive B cell'
merged.loc[memory_condition, 'B_cell_type'] = 'Memory B cell'

# 2. Extract chain type and V gene family from true_v_gene_simple
def extract_chain_and_family(gene_name):
    """Extract chain type and V gene family from gene name (e.g., IGKV1-06 -> ('Kappa', 'IGKV1'))"""
    if pd.isna(gene_name):
        return None, None
    
    gene_str = str(gene_name).upper()
    
    # Extract chain type
    if 'IGK' in gene_str:
        chain_type = 'Kappa'
    elif 'IGL' in gene_str:
        chain_type = 'Lambda'
    else:
        chain_type = None
    
    # Extract V gene family (split by '-' and take the first part)
    parts = gene_str.split('-')
    if len(parts) > 0:
        v_gene_family = parts[0]
    else:
        v_gene_family = None
    
    return chain_type, v_gene_family

# Apply the extraction function
merged[['Chain_type', 'V_gene_family']] = merged['true_v_gene_simple'].apply(
    lambda x: pd.Series(extract_chain_and_family(x))
)

# 3. Filter data - only keep sequences with valid chain types and B cell types
filtered_df = merged[
    (merged['Chain_type'].isin(['Kappa', 'Lambda'])) &
    (merged['B_cell_type'].isin(['Naive B cell', 'Memory B cell'])) &
    (merged['V_gene_family'].notna()) &
    (~merged['V_gene_family'].str.contains('H|D', na=False)) &  # Filter out heavy chain contaminants
    (~merged['V_gene_family'].str.contains('/OR', na=False))  # Filter out /OR entries
].copy()

# 4. Custom sorting function for V gene families
def sort_v_gene_families(v_gene_list):
    """Sort V gene families by their numeric part (e.g., IGKV1, IGKV2, IGKV3, etc.)"""
    import re
    
    def extract_number(gene_name):
        # Extract number from gene name (e.g., IGKV1 -> 1, IGKV10 -> 10)
        match = re.search(r'(\d+)', gene_name)
        return int(match.group(1)) if match else 0
    
    return sorted(v_gene_list, key=extract_number)

# Function to create and save individual plots
def create_plot(data, chain_type, cell_type, filename_suffix):
    plt.figure(figsize=(8, 6))
    
    # Get sorted V gene families for this data
    v_families = sort_v_gene_families(data['V_gene_family'].unique())
    
    # Create a categorical type with the sorted order
    data_copy = data.copy()
    data_copy['V_gene_family'] = pd.Categorical(data_copy['V_gene_family'], 
                                               categories=v_families, 
                                               ordered=True)
    
    # Use hue parameter to color by V_gene_family with custom palette
    sns.kdeplot(data=data_copy, x="similarity", 
                hue="V_gene_family",       # Color by V gene family
                palette=palette,           # Use custom color palette
                fill=False,               # Don't fill under the curve
                alpha=0.8,                # Line transparency
                linewidth=1.5)            # Line thickness

    plt.xlabel("Light chain sequence recovery")
    plt.ylabel("Density")
    plt.title(f"{chain_type} - {cell_type}")

    # Customize legend: rename title and remove border with thicker lines
    legend = plt.gca().get_legend()
    if legend:
        legend.set_title("V gene family")           # Rename legend title
        legend.set_frame_on(False)         # Remove legend border
        legend.set_bbox_to_anchor((1.02, 0.5))  # Adjust legend position to the right of the plot
        legend.set_loc('center left')  # Center the legend vertically
        
        # Make legend lines thicker
        for line in legend.get_lines():
            line.set_linewidth(4.0)  # Thicker legend lines

    # Remove top and right spines for cleaner look
    sns.despine()

    # Save the plot (uncomment to save)
    plt.savefig(f"/ibmm_data2/oas_database/paired_lea_tmp/paired_model/peak_analysis/bert2gpt/plots/control/{filename_suffix}_kde_transparent.png",
                dpi=600,
                bbox_inches='tight',
                transparent=True)
    plt.show()

# Create 4 separate plots

# 1. Kappa sequences from memory B cells
kappa_memory = filtered_df[
    (filtered_df['Chain_type'] == 'Kappa') &
    (filtered_df['B_cell_type'] == 'Memory B cell')
]
if not kappa_memory.empty:
    create_plot(kappa_memory, "Kappa", "Memory B cells", "kappa_memory")

# 2. Kappa sequences from naive B cells
kappa_naive = filtered_df[
    (filtered_df['Chain_type'] == 'Kappa') &
    (filtered_df['B_cell_type'] == 'Naive B cell')
]
if not kappa_naive.empty:
    create_plot(kappa_naive, "Kappa", "Naive B cells", "kappa_naive")

# 3. Lambda sequences from memory B cells
lambda_memory = filtered_df[
    (filtered_df['Chain_type'] == 'Lambda') &
    (filtered_df['B_cell_type'] == 'Memory B cell')
]
if not lambda_memory.empty:
    create_plot(lambda_memory, "Lambda", "Memory B cells", "lambda_memory")

# 4. Lambda sequences from naive B cells
lambda_naive = filtered_df[
    (filtered_df['Chain_type'] == 'Lambda') &
    (filtered_df['B_cell_type'] == 'Naive B cell')
]
if not lambda_naive.empty:
    create_plot(lambda_naive, "Lambda", "Naive B cells", "lambda_naive")

# Print summary statistics
print("Summary of data distribution:")
print(f"Total filtered samples: {len(filtered_df)}")
print(f"Kappa Memory sequences: {len(kappa_memory)}")
print(f"Kappa Naive sequences: {len(kappa_naive)}")
print(f"Lambda Memory sequences: {len(lambda_memory)}")
print(f"Lambda Naive sequences: {len(lambda_naive)}")

print("\nV Gene Families found:")
if not kappa_memory.empty:
    print(f"Kappa Memory families: {sort_v_gene_families(kappa_memory['V_gene_family'].unique())}")
if not kappa_naive.empty:
    print(f"Kappa Naive families: {sort_v_gene_families(kappa_naive['V_gene_family'].unique())}")
if not lambda_memory.empty:
    print(f"Lambda Memory families: {sort_v_gene_families(lambda_memory['V_gene_family'].unique())}")
if not lambda_naive.empty:
    print(f"Lambda Naive families: {sort_v_gene_families(lambda_naive['V_gene_family'].unique())}")

print("Plots created successfully!")
```
