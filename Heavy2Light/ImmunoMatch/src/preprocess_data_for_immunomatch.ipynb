```julia
using CSV
using DataFrames
using Plots
using StatsPlots
using Statistics
```

```julia
# bert2gpt
# run name: full_PLAbDab_healthy_human_nucleus_bert2gpt_nucleus_healthy_human_PLAbDab_max_new_tokens_115_num_epochs_50_bert_like_tokenizer_unpaired_epo_41-7 
# Average BLOSUM Score: 325.58726354968644
# Average Similarity Percentage: 62.35756522932264%
# Mean Perplexity: 2.2474937438964844

file_path = "/ibmm_data2/oas_database/paired_lea_tmp/paired_model/immuno_match/data/global_align_full_PLAbDab_healthy_human_nucleus_bert2gpt_nucleus_healthy_human_PLAbDab_max_new_tokens_115_num_epochs_50_bert_like_tokenizer_unpaired_epo_41.csv"

df = CSV.read(file_path, DataFrame);
names(df)
```

```julia
full_paired_database_file_path = "/ibmm_data2/oas_database/paired_lea_tmp/paired_model/coherence_analysis_in_oas_db/data/full_extraction_for_coherence_paired_data_header.csv"

full_paired_database_df = CSV.read(full_paired_database_file_path, DataFrame)
names(full_paired_database_df)
```

```julia
df.sequence_alignmnet_heavy_sep_light = df.input_heavy_sequence .* "[SEP]" .* df.sequence_alignment_aa_light
```

```julia
df_merged_final = innerjoin(full_paired_database_df, df, on=(:sequence_alignment_heavy_sep_light => :sequence_alignmnet_heavy_sep_light), makeunique=true)
```

```julia
# number of unique generated sequences
unique_sequences = unique(df_merged_final.generated_sequence_light)
println("Number of unique generated sequences: ", length(unique_sequences))
```

```julia
df_merged_final.locus_light = ifelse.(occursin.("K", df_merged_final.locus_light), "IGK", "IGL")
```

```julia
# save dataframe to csv
CSV.write("/ibmm_data2/oas_database/paired_lea_tmp/paired_model/immuno_match/data/all_merged_full_PLAbDab_healthy_human_nucleus_bert2gpt_nucleus_healthy_human_PLAbDab_max_new_tokens_115_num_epochs_50_bert_like_tokenizer_unpaired_epo_41-7.csv", df_merged_final)
```
