```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/ibmm_data2/oas_database/paired_lea_tmp/paired_model/unconditioned_gen_seqs/PyIR/src/bert2gpt_complete_ids_gen_true.mapping.csv')

# Remove duplicates based on generated_sequence_light column
df_unique = df.drop_duplicates(subset=['generated_sequence_light'])

# Save the result to a new CSV file
df_unique.to_csv('/ibmm_data2/oas_database/paired_lea_tmp/paired_model/unconditioned_gen_seqs/PyIR/src/bert2gpt_complete_ids_gen_true_mapping_unique.csv', index=False)

# Display info about the deduplication
print(f"Original rows: {len(df)}")
print(f"Rows after removing duplicates: {len(df_unique)}")
print(f"Duplicates removed: {len(df) - len(df_unique)}")
```

```python

# Create first FASTA file with generated_sequence_light
with open('bert2gpt_complete_ids_gen_true_mapping_unique_gen_light_seqs_heavy.fasta', 'w') as f:
    for _, row in df_unique.iterrows():
        # Write generated light sequence
        f.write(f">protein|{row['gen_fasta_id']}\n")
        f.write(f"{row['generated_sequence_light']}\n")
        
        # Write corresponding heavy sequence
        f.write(f">protein|heavy_{row['gen_fasta_id']}\n")
        f.write(f"{row['input_heavy_sequence']}\n")

# Create second FASTA file with sequence_alignment_aa_light
with open('bert2gpt_complete_ids_gen_true_mapping_unique_true_light_seqs_heavy.fasta', 'w') as f:
    for _, row in df_unique.iterrows():
        # Write aligned light sequence
        f.write(f">protein|{row['true_fasta_id']}\n")
        f.write(f"{row['sequence_alignment_aa_light']}\n")
        
        # Write corresponding heavy sequence
        f.write(f">protein|heavy_{row['true_fasta_id']}\n")
        f.write(f"{row['input_heavy_sequence']}\n")

print("FASTA files created successfully!")
print(f"File 1: generated_sequences.fasta ({len(df_unique) * 2} sequences)")
print(f"File 2: aligned_sequences.fasta ({len(df_unique) * 2} sequences)")
```
